## VGG


VGG es un modelo de red neuronal convolucional propuesto por K. Simonyan y A. Zisserman de la Universidad de Oxford en el artculo "Very Deep Convolutional Networks for Large-Scale Image Recognition". El modelo alcanza un 92,7% de precisin de las 5 pruebas principales en ImageNet  , que es un conjunto de datos de ms de 14 millones de imgenes que pertenecen a 1000 clases. En el artculo se mencioa que la red logr una precisin de prueba del 92,7% entre las 5 mejores en el conjunto de datos de ImageNet .
